# 3 Image Compression and Decompression using PCA.
In this exercise, build a PCA from scratch using explained variance method for image compression task.
You are expected to compute the necessary matrices from the scratch. Dataset: Use image of your
choice.
# 1. Load and Prepare Data:
• Fetch an image of you choice.{If colour convert to grayscale}
• Center the dataset - Standaridze the Data.
• Calculate the covaraince matrix of the Standaridze data.

```python
img = Image.open("/content/Lenna_(test_image).png").convert("L")#to grayscale

img_array = np.array(img)
print(img_array.shape)

h, w = img_array.shape
data = img_array.copy()

plt.imshow(img_array, cmap = 'gray')
plt.axis("off")
plt.show()
```

```python
mean = np.mean(data, axis=0)
#subtrating mean to center the data
centered_data = data - mean

cov_matrix = np.cov(centered_data, rowvar=False)# coverience of centered data
cov_matrix.shape
```

# 2. Eigen Decomposition and Identifying Principal Components:
• Compute Eigen Values and Eigen Vectors.

• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues.

• Identify the Principal Components with the help of cumulative Sum plot.

```python
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
sorted_indices = np.argsort(eigenvalues)[::-1] #sorted and kept in descenting order so large value comes first

eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
plt.plot(cumulative_explained_variance)
plt.xlabel("number of Components ")
plt.ylabel("cumulative explained variance")
plt.grid(True)
plt.show()

```

# 3. Reconstruction and Experiment:

• Reconstruction: Transform the original data by multiplying it with the selected eigenvectors(PCs) to obtain a lower-dimensional representation.

• Experiments: Pick Four different combination of principal components with various exlained variance value and compare the result.

• Display the Results and Evaluate.

```python
#with different k values
k_value = [10,20,50,100]  #choosing k principal component
plt.figure(figsize=(12, 8))

for i, k in enumerate(k_value):
    #selecting k principal components
    components = eigenvectors[:, :k]

    #compress data
    compressed_data = np.dot(centered_data, components)

    #decompress data
    decompressed_data = np.dot(compressed_data, components.T) + mean

    #clipvalues to be in valid grayscale range (0-255)
    decompressed_data = np.clip(decompressed_data, 0, 255)

    #compressed representation
    plt.subplot(2, 3, i + 1)
    plt.imshow(compressed_data, cmap = "gray", aspect = "auto")
    plt.title(f"Compressed Image (k={k})")
    plt.axis("off")

    #decompressed image
    plt.subplot(2, 3, i + 1) # arranging decompressed images in 2x3 grid
    plt.imshow(decompressed_data, cmap="gray")
    plt.title(f"Decompressed Image (k={k})")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
